1 Cree un diccionario que contenga los nombres  de los estudiantes y las calificaciones obtenidas por ellos en tres materias.Escriba un programa para reemplazar las calificaciones de tres materias con el total de tres materias y las calificaciones promedio.Informe también quien es el mejor de la clase.

```python
calificaciones = {
    "Luciana": {"matemáticas": 8.8, "Ingles": 9.5, "Lengua": 10.0},
    "Yon": {"matemáticas": 7.8, "Ingles": 8.8, "Lengua": 8.2}
}

# Calcular el total y el promedio de las calificaciones
for estudiante, notas in calificaciones.items():
    total = sum(notas.values())
    promedio = total / len(notas)
    calificaciones[estudiante]['Total'] = total
    calificaciones[estudiante]['Promedio'] = promedio

# Encontrar el mejor estudiante según el promedio
mejor_estudiante = max(calificaciones, key=lambda estudiante: calificaciones[estudiante]['Promedio'])

print(f"El mejor estudiante es {mejor_estudiante} con un promedio de {calificaciones[mejor_estudiante]['Promedio']:.2f}")
```

```python
# Crear un diccionario con los datos de la imagen
datos_interfaces = {
    "eth0": {"IP": "1.1.1.1", "Estado": "up"},
    "eth1": {"IP": "2.2.2.2", "Estado": "up"},
    "wlan0": {"IP": "3.3.3.3", "Estado": "down"},
    "wlan1": {"IP": "4.4.4.4", "Estado": "up"}
}

#Encontrar el estado de una interfaz determinada:
interfaz_buscar = input("Ingresa el nombre de la interfaz: ")
if interfaz_buscar in datos_interfaces:
    print(f"Estado de {interfaz_buscar}: {datos_interfaces[interfaz_buscar]['Estado']}")
else:
    print(f"La interfaz {interfaz_buscar} no existe en el diccionario.")
```

```python
#Encontrar la interfaz y la IP de todas las interfaces que están activas: 
# Recorramos el diccionario y obtengamos las interfaces activas:

interfaces_activas = [interfaz for interfaz, datos in datos_interfaces.items() if datos['Estado'] == 'up']
print(f"Interfaces activas: {', '.join(interfaces_activas)}")
```

```python
#Añadir dos nuevas entradas al diccionario:
nueva_interfaz = input("Ingresa el nombre de la nueva interfaz: ")
nueva_ip = input("Ingresa la IP de la nueva interfaz: ")
nuevo_estado = input("Ingresa el estado de la nueva interfaz (up/down): ")

datos_interfaces[nueva_interfaz] = {"IP": nueva_ip, "Estado": nuevo_estado}
print("Nuevas entradas agregadas al diccionario:")
print(datos_interfaces)
```

4. Escriba un programa en Python que imprima un diccionario donde las claves sean números en 1 y 15(ambos incluidos)y los valores sean los cuadrados  de sus claves respectivas.

```python
# Solicitar un número al usuario
numero = int(input("Ingrese un número: "))

# Crear un diccionario con las claves del 1 al 15 y sus cuadrados
diccionario_cuadrados = {i: i**2 for i in range(1, min(numero, 16))}

# Imprimir el diccionario
print(diccionario_cuadrados)
```

5. Escriba un programa en un programa en Python que ordene los elementos de una lista (contenida como valores en un diccionario)alfabéticamente. Ejemplo del Diccionario:


```python
profes = {'Math' : ['Julia', 'Miguel', 'Francisco', 'Eduardo', 'Elizabeth'],
          'Physics' : ['Antonio','Adela','Maria'],
          'Chemistry': ['Rosa','Adela']}

# Ordenar alfabéticamente las listas dentro del diccionario
for materia, nombres in profes.items():
    profes[materia] = sorted(nombres)

# Imprimir el diccionario ordenado
for materia, nombres in profes.items():
    print(f"{materia}: {nombres}")
```
