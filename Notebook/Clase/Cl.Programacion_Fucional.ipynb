Programación funcional

En la programación funcional, un problema se trata como la evaluación de una o varias 
funciones
- Las funciones pueden ser asignadas a variables y luego llamadas usando estas 
variables.
- Las funciones pueden ser pasadas como argumentos a otra función y devueltas de la 
función.
- Las funciones se pueden construir en tiempo de ejecución, de la misma forma que 
las listas, tuplas, entre otros

Ejemplo de asignación de una función a una variable y llamada a la función utilizando 
la variable

```python
def func():
    print('Hola')
def suma(x, y):
    print(x + y)
    
f = func()      # Asignación de la función a una variable
f               # Llamdo a la función 
print()
g = suma       # asignación de la función a un avariable
g(10,20)       #Llamado a suma()
```

Ejemplo de pasar una funcón como argumento a otra función:

```python
def suma1(x, y, f):
    print(x + y)
    f


def func1():
    print('Hola')
    
    
f = func1()
print()
print('#########') 
print()
suma1(10, 20, f)          
```

Ejemplo de construcción de funciones en tiempo de ejecución : Funciones Lambda

```python
def incrementar(x):
    return x + 1
resultado = incrementar(10)
print(resultado)
```

```python
incrementar = lambda x: x + 1
resultado = incrementar(10)
print(resultado)
```

Otros ejemplos de funciones lambda
 Función que calcula el cubo de un número

```python
def cubo(a):
    return a**3
b = cubo
print(b(8))
```

```python
cubo = lambda a: a**3

b = cubo
print(b(8))
```

Otros ejemplos de funciones lambda
Función que recibe 3 argumentos y devuelve el promedio de ellos

```python
def promedio(x, y, z):
    return(x + y +z)/3

print(promedio(2, 4, 11))
```

```python
promedio = lambda x, y, z:(x + y +z)/3
print(promedio(2, 4 ,11))
```
